```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np
```

##### Wczytanie danych

```python
df = pd.read_csv('data.csv', index_col=0).fillna(0)
df
```

##### Normalizacja danych

```python
scaler = StandardScaler()
scalar_fit = scaler.fit_transform(df)
```

##### Stworzenie modelu k-średnich, dla k = 4

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(scalar_fit)
df['Group'] = kmeans.predict(scalar_fit)
df
```

##### Funkcja zwracająca użytkowników z danej grupy (clustera) bez podanego jako parametr użytkownika

```python
def get_users_from_group(group, user_name):
    
    # Wypisanie wszystkich stworzonych grup
    users_group = df['Group']

    # Stworzenie tablicy wypełnionej numerem grupy podanej jako parametr o wielkości takiej jak liczba użytkowników
    group_array = np.repeat(group, 13)

    # Znalezienie indeksów tych użytkowników, których grupa jest równa wybranej grupie podanej jako parametr
    group_indices = np.where(users_group == group_array)[0]
    
    # Dane bez kolumny z grupami
    df_without_groups = df.drop(columns=['Group'])

    # Zamiana indeksów na imiona użytkowników
    users = df_without_groups.iloc[group_indices].index.tolist()

    # Usunięcie podanego jako parametr użytkownika
    users.remove(user_name)
    
    return users
```

##### Funkcja zwracająca listę rekomendowanych produktów, przyjmująca jako argument użytkowników spośród których tworzy tą listę 

```python
def get_product_from_other_users(others):
  
  # Stworzenie pustej listy rekomendowanych produktów
  recommended_products = set()

  # Dla każdego użytkownika z listy
  for user in others:

    # Pobranie oceny użytkownika
    user_ratings = df.loc[user]

    # Jeśli ocena produktu = 5, to dodaję go do listy rekomendowanych produktów
    for product, rating in user_ratings.items():
      if rating == 5:
        recommended_products.add(product)
    
    return recommended_products
```

##### Funkcja przewidująca na podstawie danych użytkownika, do której grupy będzie należał

```python
def predict_user_group(user_name):
  # Dane bez kolumny z grupami
  df_without_groups = df.drop(columns=['Group'])

  # Wczytwanie danych konkretnego użytkownika
  user_data = df_without_groups.loc[user_name, :]

  # Przerworzenie danych użytkownika do tablicy aby móc zastosować przewidywanie grupy
  user_data_array = np.array(user_data.values).reshape(1, -1)

  # Przypisanie użytkownika do jednej z grup
  predicted_group = kmeans.predict(user_data_array)
  return predicted_group
```

##### Przewidzenie grupy, do której należą członkowie naszej grupy na zajęciach (poza Asią, której nie było na zajęciach gdy uzupełnialiśmy formularz z preferencjami)

```python
Monika_predicted_group = predict_user_group('Monika')
Monika_predicted_group
```

```python
Patrycja_predicted_group = predict_user_group('Patrycja')
Patrycja_predicted_group
```

```python
Ola_predicted_group = predict_user_group('Ola')
Ola_predicted_group
```

#### Znalezienie użytkowników, którzy należą do tych samych grup

```python
users_from_Monika_group = get_users_from_group(Monika_predicted_group, 'Monika')
users_from_Monika_group
```

```python
users_from_Patrycja_group = get_users_from_group(Patrycja_predicted_group, 'Patrycja')
users_from_Patrycja_group
```

```python
users_from_Ola_group = get_users_from_group(Ola_predicted_group, 'Ola')
users_from_Ola_group
```

#### Znalezienie sugerowanych produktów, dla każdego z użytkowników

```python
Monika_recommendations = get_product_from_other_users(users_from_Monika_group)
Monika_recommendations
```

```python
Patrycja_recommendations = get_product_from_other_users(users_from_Patrycja_group)
Patrycja_recommendations
```
