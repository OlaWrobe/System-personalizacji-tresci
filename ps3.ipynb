```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

df = pd.read_csv('data.csv', index_col=0).fillna(0)
df
```

```python
scaler = StandardScaler()
scalar_fit = scaler.fit_transform(df)
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(scalar_fit)
df['Cluster'] = kmeans.predict(scalar_fit)
df
```
